## **`Problem 2: Bayesian Inference Calculation`**

Let's calculate the probability of having Pneumonia given abnormal chest X-ray results.

Given:
- $P(P)$ = Probability of having Pneumonia = `0.02` *(2% of the population)*
- $P(X|P)$ = Probability of abnormal X-ray given Pneumonia = `0.90`
- $P(X|¬P)$ = Probability of abnormal X-ray given no Pneumonia = `0.05`

 We want to calculate P(P|X) using Bayes' Theorem:

$$P(P|X) = \frac{P(X|P) \times P(P)}{P(X)}$$

## *`Step 1: Calculate P(X):`*
$$P(X) = P(X|P) \times P(P) + P(X|¬P) \times P(¬P)$$
$$P(X) = 0.90 \times 0.02 + 0.05 \times 0.98 = 0.018 + 0.049 = 0.067$$

## *`Step 2: Apply Bayes' Theorem:`*
$$P(P|X) = \frac{0.90 \times 0.02}{0.067} \approx 0.2687$$

Therefore, the probability of having Pneumonia given abnormal chest X-ray results is approximately *`26.87%`*.

## **`Problem 3: Decision-Making Simulation`**

Let me simulate a decision-making process for recommending hospitalization based on the calculated probability.

Assume:
- Hospitalization costs `$1000` per day
- Average hospital stay for pneumonia is `5` days
- Not hospitalizing a pneumonia patient may lead to complications costing `$10000` on average
- We recommend hospitalization if the expected value of hospitalizing is lower than not hospitalizing

```python
p_pneumonia = 0.2687  # Probability of pneumonia given abnormal X-ray
cost_hospitalization = 1000 * 5  # $1000 per day for 5 days
cost_complications = 10000

expected_cost_hospitalize = cost_hospitalization
expected_cost_no_hospitalize = p_pneumonia * cost_complications

print(f"Expected Cost of Hospitalization: ${expected_cost_hospitalize:.2f}")
print(f"Expected Cost of No Hospitalization: ${expected_cost_no_hospitalize:.2f}")

if expected_cost_hospitalize < expected_cost_no_hospitalize:
    print("Recommendation: Proceed with hospitalization")
else:
    print("Recommendation: No hospitalization required")
  
```

````python
#@title # **`Problem #4: Visualization of Probability Distribution`**
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

#@markdown ## Probability values:
#@markdown - $ n = \text{number of people} $
#@markdown - $ p = \text{probability of having Pneumoinia} $
n = 1000  #@param {type:"number"}
p = 0.02  #@param {type:"number"}

# Calculate probabilities for 0 to 50 cases (unlikely to have more in 1000 people)
k = np.arange(0, 51)
probabilities = binom.pmf(k, n, p)

# Plot the distribution
plt.figure(figsize=(12, 6))
plt.bar(k, probabilities)
plt.title('Binomial Distribution of Pneumonia Cases in 1000 People')
plt.xlabel('Number of Pneumonia Cases')
plt.ylabel('Probability')
plt.xticks(np.arange(0, 51, 5))
plt.show()
````
